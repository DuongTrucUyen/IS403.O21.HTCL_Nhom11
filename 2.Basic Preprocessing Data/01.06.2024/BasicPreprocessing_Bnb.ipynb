```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Đọc tệp CSV và lưu dữ liệu vào một DataFrame
df = pd.read_csv('/content/Bnb.csv')
df = df[['Close']]
#In ra nội dung của DataFrame
print(df)
```

```python
#In ra các thống kê mô tả của DataFrame
print(df.describe())
```

```python
#Kiểm tra missing data
df.isna().sum()
```

```python
# Vẽ biểu đồ histogram
plt.hist(df)
# Hiển thị biểu đồ
plt.show()
```

```python
#Vẽ biểu đồ box plot
plt.boxplot(df)
#Hiển thị biểu đồ
plt.show()
```
